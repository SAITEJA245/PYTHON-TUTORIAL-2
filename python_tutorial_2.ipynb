```python
for i in range(5):
  if i==3:
    break  # break skips the whole loop itself so from i=3 nothing will be printed
  print("Hello ",i)
```

```python
for i in range(5):
  if i==3:
    continue  # when used this function the below print will not be executed and goes to the begining and starts with i=4
  print("Hello ",i)
```

**when we assign b=a the array will be aliased because the same array has two names a and b we call b is alised to a**

**veiw()** is a function which helps us to create a new array at a different location in the memory

when we use veiw  function it does **SHALLOW COPY** which means altough a new address is created for the array c it is still connected to the array a so if you change the value of a element in array c it will also be changed in array a to avoid this from happening we use **DEEP COPY**

in deep copy we use the function **copy()** when we use c=a.copy() the address of array a and array c is different and also when you commit changes to the elements in array c it doesn't change the elements in array a which means they are not connected like they were when we used veiw() function





```python
# copying array
from numpy import *
a=array([2,4,9,1,16])
b=a      # instead of copying elements from a to b if we assign b=a it will just make point b to the same address it doesn't create a new array in the memory as here we have two names to the same array in the memory this is called ALIASING
print(id(a),id(b))  # as a new array is not created in the array we can see that their address are the same
c = a.view()   # veiw is a function which helps us to create a new array at a different location in the memory
print(c,id(c))   # check that the array c is having a different address
c[1] = 10   # since veiw() function does shallow copying we can see that when we change c[1] then a[1] will also be changed
print(c)
print(a)
```

```python
a=array([2,4,9,1,16])
c=a.copy()
print(a,id(a))
print(c,id(c))
c[1]=22
print(a)
print(c)   # as we used copy() to do deep copy we can see that the elementa of the array a are unchanged when the elements in array c are changed
```

GLOBAL and LOCAL VARIABLES

local variables are the avariables which are defined inside the function and global are the ones outside

global variables can be accessed by all functions 

local variables can only be accessed inside the function in which they are declared

```python
a=10
def test():
  a=20      # inside a function preferance will be given to local variable over global variable
  print(a ,'in function')
test()
print(a)
```

```python
a=10
def test():
# we can access a global variable but the moment we assign it a value it will create a new local variable 
# so in order to change the value of a use global keyword which says to update global a while running
  global a  
  a=20      
  print(a ,'in function')  # here we can not use a as another local variable
test()
print(a,'outside function')
```

```python
#fibonacci sequence
n=int(input())
l=[0,1]
def fib(n):
  i=2
  while i<n:
    l.append(l[i-1]+l[i-2])
    i+=1
  return l
print(fib(n))
```

```python
# code to print fibonacci numbers below given input bound
def fibo(bound):
    a, b = 0, 1
    while a < bound:
        print(a, end="  ")
        a, b = b, a+b
bound=int(input())
fib(bound)
```

```python
def factorial(n):
  r=1
  for i in range (2,n+1):
    r=r*i
  return r
print(factorial(10))
```

```python
# using recursion for factorial
def fact(n):
  if n==0:
    return 1
  return n*fact(n-1)
print(fact(10))
```
